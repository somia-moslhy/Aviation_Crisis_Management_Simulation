# 📖 Data Dictionary

This section outlines the exact schema of the datasets used for our Aviation Crisis Simulation.

### 1. `airlines.csv` (Airlines Reference)
* **IATA_CODE**: Unique 2-letter identifier assigned to the airline.
* **AIRLINE**: The full commercial name of the airline.

### 2. `airports.csv` (Airports Reference)
* **IATA_CODE**: Unique 3-letter identifier for the airport.
* **AIRPORT**: The full official name of the airport.
* **CITY / STATE / COUNTRY**: Geographical location details.
* **LATITUDE / LONGITUDE**: GPS coordinates of the airport (Useful for geospatial mapping).

### 3. `flights.csv` (Core Operational Data)
**📅 Date & Identification:**
* **YEAR / MONTH / DAY / DAY_OF_WEEK**: Date attributes of the flight.
* **AIRLINE**: The 2-letter code of the operating airline.
* **FLIGHT_NUMBER**: The specific flight identifier.
* **TAIL_NUMBER**: The unique registration number of the aircraft.
* **ORIGIN_AIRPORT**: The 3-letter code of the departure airport.
* **DESTINATION_AIRPORT**: The 3-letter code of the intended arrival airport.
* **DISTANCE**: Flight distance between origin and destination (in miles).

**🛫 Departure Metrics:**
* **SCHEDULED_DEPARTURE**: Planned departure time.
* **DEPARTURE_TIME**: Actual departure time.
* **DEPARTURE_DELAY**: Difference in minutes (Actual - Scheduled).
* **TAXI_OUT**: Time elapsed (in minutes) between leaving the origin gate and wheels taking off.
* **WHEELS_OFF**: The exact time the aircraft's wheels left the ground.

**⏱️ Flight Duration:**
* **SCHEDULED_TIME**: Planned total flight duration.
* **ELAPSED_TIME**: Actual total time from gate to gate.
* **AIR_TIME**: Actual time spent in the air.

**🛬 Arrival Metrics:**
* **WHEELS_ON**: The exact time the aircraft's wheels touched the ground.
* **TAXI_IN**: Time elapsed (in minutes) between wheels landing and arriving at the destination gate.
* **SCHEDULED_ARRIVAL**: Planned arrival time.
* **ARRIVAL_TIME**: Actual arrival time.
* **ARRIVAL_DELAY**: Difference in minutes (Actual - Scheduled).

**🚨 Crisis & Disruption Indicators (Key Features):**
* **DIVERTED**: Binary indicator (1 = Yes, 0 = No). The flight landed at an alternative airport.
* **CANCELLED**: Binary indicator (1 = Yes, 0 = No). The flight was completely canceled.
* **CANCELLATION_REASON**: Categorical code for cancellation.

**⏳ Delay Breakdown (in minutes):**
* **AIR_SYSTEM_DELAY**: Delay caused by the national aviation system (e.g., heavy traffic).
* **SECURITY_DELAY**: Delay caused by security breaches or protocols.
* **AIRLINE_DELAY**: Delay caused by the carrier (e.g., crew issues, maintenance).
* **LATE_AIRCRAFT_DELAY**: Delay caused by the late arrival of the same aircraft from a previous flight.
* **WEATHER_DELAY**: Delay caused by extreme weather conditions.
